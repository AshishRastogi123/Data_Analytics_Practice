# Outlier Detection

```python
# to detection we plot >> distribution , boxplot, violenplot
```

```python
students_grade_df=df.copy()
students_grade_df
```

```python
students_grade_df=students_grade_df.drop(columns=['G1','G2','G3'], axis=1)
students_grade_df
```

```python
#Q1-1.5*IQR
#Q3+1.5*IQR
```

```python
def detect_outlier(columns):
  outlier_indices=[]
  for col in columns:
    Q1=students_grade_df[col].quantile(0.25)
    Q3=students_grade_df[col].quantile(0.75)
    IQR=Q3-Q1

    # make condition
    mask= (students_grade_df[col]>= (Q1-1.5*IQR)) & (students_grade_df[col]<= (Q3 + 1.5*IQR))
    mask= mask.to_numpy()

    #find the indices of values that do not satisfy the mask condition>> that will be outlier
    false_indices=np.argwhere(~mask)
    outlier_indices.append(false_indices)
  return np.unique(np.concatenate(outlier_indices).ravel())


#np.concatenate(outlier_indices)>> combine all the outlier index arrays into a single array
#so each column will have one array
#.ravel>> flattens the multi-dimensional array into 1 d array



```

```python
num_cols=['age', 'absences']
detect_outlier(num_cols)
```
